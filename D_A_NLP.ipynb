```python
import re
# Function to extract numeric value from EUR strings
def extract_eur_value(value):
    if isinstance(value, str):  # Ensure value is a string
        numbers = re.findall(r'\d+', value)  # Extract numbers
        return int(''.join(numbers)) if numbers else None  # Convert to int
    return None  # Return None for non-string values

# Apply function to list
m_growth = [extract_eur_value(value) for value in m_growth]

print(m_growth)
```

```python
m_stable = [extract_eur_value(value) for value in m_stable]
m_risk=[extract_eur_value(value) for value in m_risk]

print(m_stable)
print(m_risk)
```
